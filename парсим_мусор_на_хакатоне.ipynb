```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

response = requests.get('http://fcao.ru/groro')
soup = BeautifulSoup(response.text)

soup
```

```python
data_per_page = soup
list_per_page = []
for tag in data_per_page:    
    region = soup.find('td', attrs = {'class': 'views-field views-field-field-region'}).get_text().strip()
    name = soup.find('td', attrs = {'class': 'views-field views-field-field-object'}).get_text().strip()
    purpose = soup.find('td', attrs = {'class': 'views-field views-field-field-nazn'}).get_text().strip()
    impact = soup.find('td', attrs = {'class': 'views-field views-field-field-negativ'}).get_text().strip()
list_per_data = [region, name, purpose, impact]
list_per_page.append(list_per_data)
    
list_per_page
```

```python
list_all = []
for letter in range(0, 428):
    soup = BeautifulSoup(requests.get("http://fcao.ru/groro?page="+str(letter)).text)
    list_per_page = []
    for i in range(0, int(len(soup.find_all('td', attrs = {'class': 'views-field views-field-field-region'})))):
        region = soup.find_all('td', attrs = {'class': 'views-field views-field-field-region'})[i].get_text().strip()
        name = soup.find_all('td', attrs = {'class': 'views-field views-field-field-object'})[i].get_text().strip()
        purpose = soup.find_all('td', attrs = {'class': 'views-field views-field-field-nazn'})[i].get_text().strip()
        impact = soup.find_all('td', attrs = {'class': 'views-field views-field-field-negativ'})[i].get_text().strip()
        list_per_string = [region, name, purpose, impact]   
        list_per_page.append(list_per_string)
    list_all.extend(list_per_page)
list_all
```

```python
data = pd.DataFrame(list_all, columns = ["Регион", "Название объекта", "Тип объекта", "Негативное воздействие на окружающую среду"])
data
```
